```python
import pandas as pd
import numpy as np
from prophet import Prophet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import joblib

# Step 1: Load Data from CSV
file_path = '/Users/noraizabid/Desktop/Dissertation Algos/prophet/BTCUSD_1h_2021_to_2024_cleaned.csv'  # Update with your actual file path
data = pd.read_csv(file_path)

# Convert the time column to datetime
data['time'] = pd.to_datetime(data['time'])  # Letting pandas infer the datetime format
data = data.sort_values('time')

# Continue with the rest of your code...


# Prepare the data for Prophet
df = data[['time', 'close', 'open', 'high', 'low', 'volume']].rename(columns={'time': 'ds', 'close': 'y'})

# Step 2: Train-Test Split (75% for training)
train_df, test_df = train_test_split(df, train_size=0.75, shuffle=False, random_state=42)

# Step 3: Prophet Model Training on 75% of the data
model = Prophet()

# Add all additional features as regressors
model.add_regressor('open')
model.add_regressor('high')
model.add_regressor('low')
model.add_regressor('volume')

# Train the model on the training data (75% of the total data)
model.fit(train_df)

# Step 4: Refine Prediction Using Last 7 Days
# Filter the last 7 days of data
last_7_days_df = df[df['ds'] >= df['ds'].max() - pd.Timedelta(days=7)]

# Predict the next 1 hour using the model and last 7 days data
future = model.make_future_dataframe(periods=1, freq='H')

# Add the regressor columns to the future dataframe using the last 7 days data
future['open'] = last_7_days_df['open'].iloc[-len(future):].reset_index(drop=True)
future['high'] = last_7_days_df['high'].iloc[-len(future):].reset_index(drop=True)
future['low'] = last_7_days_df['low'].iloc[-len(future):].reset_index(drop=True)
future['volume'] = last_7_days_df['volume'].iloc[-len(future):].reset_index(drop=True)

forecast = model.predict(future)

# Save the model using joblib
model_filename = 'prophet_model_75_percent_training_last_7_days_refinement.pkl'
joblib.dump(model, model_filename)

# Step 5: Evaluation and Visualization
# Performance metrics
y_true = test_df['y']
y_pred = forecast['yhat'].iloc[-len(test_df):]

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(train_df['ds'], train_df['y'], label='Train Data')
plt.plot(test_df['ds'], test_df['y'], label='Test Data')
plt.plot(forecast['ds'], forecast['yhat'], label='Predicted Data')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
